How are temperature values distributed?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for Rainfall	
plt.figure(figsize=(8, 5))
sns.histplot(df['MaxTemp'], kde=True)
plt.title("MaxTemp Distribution")
plt.xlabel("MaxTemp(Â°C)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for Rainfall	
plt.figure(figsize=(8, 5))
sns.histplot(df['MinTemp'], kde=True)
plt.title("MinTemp Distribution")
plt.xlabel("MinTemp(Â°C)")
plt.show()
```

Is there a seasonal pattern in temperature, rainfall, or other variables?

```python
df['Date'] = pd.to_datetime(df['Date'])
df['month'] = df['Date'].dt.month

# Average temperature by month
monthly_avg = df.groupby('month')['MaxTemp'].mean()

plt.figure(figsize=(10, 6))
monthly_avg.plot(kind='bar')
plt.title("Average Monthly Temperature")
plt.xlabel("Month")
plt.ylabel("Maximum Temperature (Â°C)")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['month'] = df['Date'].dt.month

# Average temperature by month
monthly_avg = df.groupby('month')['MinTemp'].mean()

plt.figure(figsize=(10, 6))
monthly_avg.plot(kind='bar')
plt.title("Average Monthly MinTemp")
plt.xlabel("Month")
plt.ylabel("Minimum Temperature (Â°C)")
plt.show()
```

What is the trend of maximum and minimum temperatures over time?

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['MaxTemp'], label="Max Temp")
plt.plot(df['Date'], df['MinTemp'], label="Min Temp")
plt.title("Max and Min Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (Â°C)")
plt.legend()
plt.show()
```

Is there a relationship between temperature and humidity?

```python
# Scatter plot for temperature vs humidity
plt.figure(figsize=(8, 6))
sns.scatterplot(x='temperature', y='humidity', data=df)
plt.title("Temperature vs Humidity")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Humidity (%)")
plt.show()
```
